```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

chemicals_data = pd.read_csv('capdata.csv')

X = chemicals_data.drop(columns=['Concentration'])
y = chemicals_data['Concentration']

model = DecisionTreeClassifier()
model.fit(X , y)

predicitions = model.predict([ [10,24,48] , [4,23,93] ])
predicitions
```
